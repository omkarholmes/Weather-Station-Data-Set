# Clustering

```python
from sklearn.cluster import DBSCAN
```

```python
X=dataset[dataset.columns[1:6]]
X
```

```python
from sklearn.preprocessing import StandardScaler
stscaler = StandardScaler()
X[X.columns[0:2]] = stscaler.fit_transform(X[X.columns[0:2]])
```

```python
data =X[X.columns[2:5]]
data
```

```python
from sklearn.preprocessing import StandardScaler
stscaler = StandardScaler()
data = stscaler.fit_transform(data)
data
```

```python
dbsc = DBSCAN(eps = 0.2, min_samples=15).fit(data)
labels = dbsc.labels_
core_samples = np.zeros_like(labels,dtype=bool)
core_samples[dbsc.core_sample_indices_]=True
```

```python
set(labels)
```

```python
X["Clust_ds"]=dbsc.labels_
```

```python
dataset=X
group =dataset.groupby("Clust_ds")
group.count()
```

```python
import matplotlib.pyplot as plt
import folium
```

```python
map = folium.Map(location=[dataset["Lat"].agg(np.mean),dataset["Long"].agg(np.mean)], zoom_start=4)
colors=["yellow","red","blue","green","cyan","magenta","black"]
for location in range(0,dataset.shape[0]):
        row=dataset.iloc[location]
        folium.CircleMarker(location=[row["Lat"],row["Long"]],
                    color=colors[int(row["Clust_ds"]+1)],
                    radius= 2,
                    fill=True,
                    fill_color="purple",
                    fill_opacity=0.6).add_to(map)
map
```
